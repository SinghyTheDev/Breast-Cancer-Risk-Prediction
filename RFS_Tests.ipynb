```python
import pandas as pd
```

```python
#Import function from RFS file to generate models
from RFS_Prediction import best_eval
```

```python
#Change text for .xls to input data
inputData = 'testDataset.xls'
```

```python
pred = best_eval(inputData)
```

```python
#Printing values to ascertain correct format
predicted = pred.flatten()
print(predicted)
```

```python
#Creates simple output xls file with ID, RFS
import xlwt
workbook = xlwt.Workbook()
worksheet = workbook.add_sheet('Sheet 1')

ids=pd.read_excel(inputData, index_col=False)
ids= ids["ID"]

ids = ids.to_numpy()

index = 0;
for i in ids: 
    worksheet.write(index, 0, i)
    worksheet.write(index, 1, int(predicted[index]))
    index = index +1 

workbook.save('rfs_outcome.xls')
```
